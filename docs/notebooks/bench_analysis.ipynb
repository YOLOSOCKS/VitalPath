# Aegis Benchmark Analysis (Jupyter)

This notebook reads the latest `docs/bench/results_*.jsonl` file and generates figures in `docs/figures/`.

**Prereqs**: run the backend, then run `backend/bench/run_bench.py` at least once.


```python
from pathlib import Path
import sys

# Import the figure generator module
bench_dir = Path('..') / 'bench'
sys.path.append(str(bench_dir.resolve()))
import make_figures

results = make_figures.find_latest_results(Path('..') / 'bench')
results
```

```python
if results is None:
    raise RuntimeError('No results_*.jsonl found in docs/bench. Run backend/bench/run_bench.py first.')

summary = make_figures.generate_all(results, out_dir=Path('..') / 'figures', theme='dark')
summary
```

## Output

- Figures are written to `docs/figures/`
- A markdown summary is written to `docs/bench/` (same folder as results)
